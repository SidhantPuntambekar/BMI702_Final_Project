```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('GDPa1_v1.2_20250814.csv')
print('Shape:', df.shape)
print('Columns:', list(df.columns))
```

```python
features = ['HIC', 'SMAC', 'HAC', 'PR_CHO', 'PR_Ova', 'SEC %Monomer',
            'AC-SINS_pH6.0', 'AC-SINS_pH7.4', 'Tm1', 'Tm2', 'Titer']

missing_all = df[features].isna().sum().rename('Missing (n=246)')
print(missing_all.to_string())
```

```python
print('highest_clinical_trial_asof_feb2025 value counts:')
print(df['highest_clinical_trial_asof_feb2025'].value_counts())
print()
print('est_status_asof_feb2025 value counts:')
print(df['est_status_asof_feb2025'].value_counts())
```

```python
approved = df[df['highest_clinical_trial_asof_feb2025'].isin(['Approved', 'Approved (withdrawn)'])]
not_approved = df[df['est_status_asof_feb2025'] == 'Discontinued']

print('Approved count:    ', len(approved))
print('Not Approved count:', len(not_approved))

subset = df[
    df['highest_clinical_trial_asof_feb2025'].isin(['Approved', 'Approved (withdrawn)']) |
    (df['est_status_asof_feb2025'] == 'Discontinued')
]
print('Subset total:      ', len(subset))
```

```python
missing_subset = subset[features].isna().sum().rename('Missing (n=198)')
pct_subset = (subset[features].isna().mean() * 100).round(1).rename('Missing %')

summary = pd.concat([missing_subset, pct_subset], axis=1)
print(summary.to_string())
```

```python
comparison = pd.DataFrame({
    'Missing (n=246)': df[features].isna().sum(),
    'Missing (n=198)': subset[features].isna().sum(),
})
print(comparison.to_string())
```
